```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import math
import random
```

```python
dataset = pd.read_csv("C:/Users/divya/AIML/Cr-02-ML_AZ/MLAZ_Course/datasets/Ads_CTR_Optimisation.csv")

dataset.head()
```

```python
N = len(dataset)
d = len(dataset.columns)
ads_selected = []
number_of_rewards1 = [0] * d
numbers_of_rewards0 = [0] * d
total_rewards = 0
```

```python
for n in range(0, N):
    ad = 0
    max_random = 0
    for i in range(0,d):
        random_beta = random.betavariate(number_of_rewards1[i] + 1, numbers_of_rewards0[i] +1)
        if random_beta > max_random:
            max_random = random_beta
            ad = i
    ads_selected.append(ad)
    reward = dataset.values[n, ad]
    if reward == 1:
        number_of_rewards1[ad] = number_of_rewards1[ad] + 1
    else:
        numbers_of_rewards0[ad] = numbers_of_rewards0[ad] + 1

    total_rewards = total_rewards + reward
```

```python
plt.hist(ads_selected)
plt.title('Histogram of ads selections')
plt.xlabel('Ads')
plt.ylabel('Number of times each ad was selected')
plt.show()
```
