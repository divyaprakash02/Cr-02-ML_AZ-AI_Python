```python
import pandas as pd
import numpy as np
```

```python
X = df.drop(['Position','Salary'], axis=1).values
y = df['Salary'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)#not needed 
```

```python
from sklearn.linear_model import LinearRegression
linerRegression = LinearRegression()
```

```python
print('using linear regression')
```

```python
linerRegression.fit(X_train, y_train)#not needed
```

```python
linerRegression.fit(X, y)
```

```python
y_lin_predict = linerRegression.predict(X_test)
```

```python
print(y_lin_predict)
```

```python
#not needed
np.set_printoptions(precision=2)
print(np.concatenate((y_lin_predict.reshape(len(y_lin_predict),1), y_test.reshape(len(y_test),1)),axis=1))
```

```python
print('using polynomial regression')
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly_reg = PolynomialFeatures(degree=2)#this value is changeable at degree of 2
X_poly = poly_reg.fit_transform(X)
line_reg_2 = LinearRegression()
line_reg_2.fit(X_poly, y)
```

```python
import matplotlib.pyplot as plt
```

```python
print(X)
print(linerRegression.predict(X))
```

```python
#viz linear regression
plt.scatter(X, y, color= 'red')
plt.plot(X, linerRegression.predict(X), color='blue')
plt.title('Truth or Bluff (Linear Regression)')
plt.xlabel('Position Salary')
plt.ylabel('Salary')
plt.show()
```

```python
#viz polynomial regression
plt.scatter(X, y, color= 'red')
plt.plot(X, line_reg_2.predict(X_poly), color='blue')
plt.title('Truth or Bluff (Polynomial Regression)')
plt.xlabel('Position Salary')
plt.ylabel('Salary')
plt.show()
```

```python
poly_reg = PolynomialFeatures(degree=4)#this value is changeable at degree of 4
X_poly = poly_reg.fit_transform(X)
line_reg_2 = LinearRegression()
line_reg_2.fit(X_poly, y)

plt.scatter(X, y, color= 'red')
plt.plot(X, line_reg_2.predict(X_poly), color='blue')
plt.title('Truth or Bluff (Polynomial Regression)')
plt.xlabel('Position Salary')
plt.ylabel('Salary')
plt.show()
```
