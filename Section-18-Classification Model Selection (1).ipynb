```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
df = pd.DataFrame(pd.read_csv("C:/Users/divya/AIML/Cr-02-ML_AZ/MLAZ_Course/datasets/ClassificationData.csv"))
```

```python
df.head()
```

```python
df.describe()
```

```python
X = df.drop('Class', axis=1).values
y = df['Class'].values
```

```python
print(y)
```

```python
print(X)
```

```python
#splitting dataset into training and test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
#feature scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

# 1. Decision Tree Classification

```python
from sklearn.tree import DecisionTreeClassifier
dcClassfier = DecisionTreeClassifier(random_state=0, criterion='entropy')
```

```python
dcClassfier.fit(X_train, y_train)
```

```python
y_dc_pred = dcClassfier.predict(X_test)
```

```python
print(classification_report(y_test, y_dc_pred))
print()
print(confusion_matrix(y_test, y_dc_pred))
print()
print(accuracy_score(y_test, y_dc_pred))
```

# 2. KNN Classification

```python
from sklearn.neighbors import KNeighborsClassifier
ngClassifier = KNeighborsClassifier(n_neighbors=5,metric='minkowski',p=2)
```

```python
ngClassifier.fit(X_train, y_train)
```

```python
y_knn_pred = ngClassifier.predict(X_test)
```

```python
print(classification_report(y_test, y_knn_pred))
print()
print(confusion_matrix(y_test, y_knn_pred))
print()
print(accuracy_score(y_test, y_knn_pred))
```

# 3. Kernel SVM

```python
from sklearn.svm import SVC
ksvm_Classifier = SVC(kernel='rbf', random_state=0)
```

```python
ksvm_Classifier.fit(X_train, y_train)
```

```python
y_ksvm_pred = ksvm_Classifier.predict(X_test)
```

```python
print(classification_report(y_test, y_ksvm_pred))
print()
print(confusion_matrix(y_test, y_ksvm_pred))
print()
print(accuracy_score(y_test, y_ksvm_pred))
```

# 4. Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lr_Classifier = LogisticRegression()
```

```python
lr_Classifier.fit(X_train, y_train)
```

```python
y_lr_pred = lr_Classifier.predict(X_test)
```

```python
print(classification_report(y_test, y_lr_pred))
print()
print(confusion_matrix(y_test, y_lr_pred))
print()
print(accuracy_score(y_test, y_lr_pred))
```
